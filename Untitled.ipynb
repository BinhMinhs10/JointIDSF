```python
train_word = 'data/vi-atis-fix/train/word/seq.out'
dev_word = 'data/vi-atis-fix/dev/word/seq.out'
test_word = 'data/vi-atis-fix/test/word/seq.out'
```

```python
train_syl = 'data/vi-atis-fix/train/syllable/seq.out'
dev_syl = 'data/vi-atis-fix/dev/syllable/seq.out'
test_syl = 'data/vi-atis-fix/test/syllable/seq.out'
```

```python
set_1 = set()
```

```python
set_2 = set()
```

```python
list_1 = []
list_2 = []
```

```python
with open(train_word) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_1.add(tok)
        list_1.append(sent_)
with open(dev_word) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_1.add(tok)            

        list_1.append(sent_)
with open(test_word) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_1.add(tok)            
        list_1.append(sent_)
```

```python
with open(train_syl) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_2.add(tok)

        list_2.append(sent_)
with open(dev_syl) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_2.add(tok)            

        list_2.append(sent_)
with open(test_syl) as f:
    for line in f:
        sent_ = []
        for tok in line.split():
            if 'B-' in tok:
                sent_.append(tok)
            set_2.add(tok)            
        list_2.append(sent_)
```
